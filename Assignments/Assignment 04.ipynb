Q4 — Count Frequency of Numbers
numbers = [1,2,2,3,3,3,4,4,4,4]
Create frequency dictionary

```python
number = [1,2,2,3,3,3,4,4,4,4]
frequency = {}
for num in number:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

Q5 — Word Counter
Take sentence input and count word frequency

```python
sentence = input("Enter a sentence: ")
word_count ={}
for word in sentence.split():
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print(word_count)
```

Q6 — Student Marks & Average
marks = {"Math":80, "English":70, "Physics":90}
Calculate total and average

```python
marks = {
    "Math": 80,
    "English": 70,
    "Physics": 90
}
total_marks = sum(marks.values())
average_marks = total_marks / len(marks)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
```

Q7 — Highest Scoring Student                                                                                                                  
students = {
"Ali":85,
"Sara":92,
"Ahmed":78,
"Zara":95
}                                                                                                                                             
Find highest scoring student

```python
students = {
    "ali": 85,
    "sara": 92,
    "ahmed": 78,
    "zara": 95
}
highest_score = max(students.values())
for student, score in students.items():
    if score == highest_score:
        print(student, ":", score)

```
